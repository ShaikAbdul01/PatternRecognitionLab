```python
array = df.values
x = array[:,0:13]
y = array[:,13]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=11)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=10000),
    'Naive Bayes': GaussianNB(),
    'Linear Discriminant Analysis': LinearDiscriminantAnalysis(),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=11),
    'Decision Tree': DecisionTreeClassifier(random_state=11),
    'SVC (Linear Kernel)': SVC(kernel='linear', random_state=11),
    'SVM (RBF Kernel)': SVC(kernel='rbf', random_state=11),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5)
}
```

```python
results = []
for name, model in models.items():
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    acc = accuracy_score(y_test, predictions)
    results.append((name, acc))
    print(f"\n{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, predictions))
```

```python
model_names = [r[0] for r in results]
accuracies = [r[1] for r in results]
model_names,accuracies
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=accuracies, y=model_names, color='skyblue')
plt.xlabel('Accuracy')
plt.title('Comparison of ML Models on Heart Disease Dataset')
plt.xlim(0, 1)
plt.savefig('accuracy.png', dpi=800)
plt.show()
```
